```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
#generate synthetic data
data = np.random.normal(0, 1, (1000, 20))

#Introdue some anomalies
anomalies = np.random.normal(5, 1,(50, 20))
data_with_anomalies = np.vstack([data , anomalies])
```

```python
data_with_anomalies.shape
```

```python
sns.kdeplot(data)
```

```python
#normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data_with_anomalies)

#split the data into training and testing sets
x_train, x_test = train_test_split(scaled_data, test_size=0.2, random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#define the input layer
input_layer = Input(shape=(x_train.shape[1],))

#encoder layers
encoded = Dense(16, activation = 'relu')(input_layer)
encoded = Dense(8, activation = 'relu')(encoded)
encoded = Dense(4, activation ='relu')(encoded)

#decoded layers
decoded = Dense(8, activation = 'relu')(encoded)
decoded = Dense(16, activation = 'relu')(decoded)
decoded = Dense(x_train.shape[1], activation ='sigmoid')(decoded)

#autoencoder
autoencoder = Model(inputs = input_layer, outputs = decoded)
autoencoder.compile(loss = 'mse', optimizer ='adam')
```

```python
autoencoder.compile(loss = 'mse', optimizer = 'adam', metrics = ['accuracy'])
```

```python
history = autoencoder.fit(x_train, x_train, epochs =50, batch_size =10, shuffle = True, validation_split = 0.1)
```

```python
#predict reconstruction on test data
x_test_pred = autoencoder.predict(x_test)
```

```python
#calculate the mean squared error
mse = mean_squared_error(x_test, x_test_pred, multioutput = 'raw_values')
mse
```

```python
#Define threshold for anomaly detection
threshold = np.percentile(mse, 90)
threshold
```

```python
#Identify anomalies
anomalies = mse>threshold
print(f'Number of anomalies detected: {np.sum(anomalies)}')
```

```python
np.mean((x_test-x_test_pred)**2, axis = 0)
```

```python
sum(mse > threshold)
```

```python
import matplotlib.pyplot as plt

#plot the loss over epochs
plt.plot(history.history['loss'], label ='training loss')
plt.plot(history.history['loss'], label = 'validation loss')
plt.legend()
plt.show()

#plot the histogram
plt.hist(mse, bins=20)
plt.axvline(threshold, color= 'green', linestyle = '--')
plt.show()
```
